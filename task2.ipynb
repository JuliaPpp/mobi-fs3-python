## Intensity clipping

First we define some fixed thresholds $T_1$, $T_2$:

```python
T1 = 0.31
T2 = 0.59
```

### Using `ndarray.clip` method

**<a id="#">Task 2.3.1.</a>** Enter your solution below:

```python
bild2=img.copy()
bild2_clip=bild2.clip(T1, T2)
plt.imshow(bild2_clip, 'gray')
```

### Not using the `ndarray.clip` method

**<a id="#">Task 2.3.2 (a).</a>** Enter your solution below:

```python
img_2=img.copy()
for i in range(img_2.shape[0]):
    for j in range(img_2.shape[1]):
        if img_2[i, j] < T1:
            img_2[i, j] = T1
        elif img_2[i, j] > T2:
            img_2[i, j] =T2

plt.figure(figsize=(15,8))
plt.imshow(img_2, 'gray')
plt.colorbar()
```

**<a id="#">Task 2.3.2 (b).</a>** Enter your solution below:

```python
img_3=img.copy()

for i in numpy.ndindex(img_3.shape):
    if img_3[i] < T1:
        img_3[i] = T1
    elif img_3[i] > T2:
        img_3[i] = T2

plt.figure(figsize=(15, 8))
plt.imshow(img_3, 'gray')
plt.colorbar()
```
